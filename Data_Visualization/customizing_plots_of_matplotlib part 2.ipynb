# Customizing plots in seaborn and matplotlib

Loading Libraries and Data 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Diamond dataset 

```python
df=pd.read_csv('D:\Data Visualization\Data_Visualization\diamonds.csv')
df.head()
```

# Why we use sampling in our large dataset 

Sampling is a powerful tool that balances efficiency, scalability, and representativeness in data analysis. It allows for meaningful insights and model development without requiring exhaustive computations on the full dataset.

```python
print('Number of Rows before taking sampling:',df.shape[0])# number of rows 
print('Number of Columns before taking sampling:',df.shape[1]) # number of columns
print("--------------------------------------")

df=df.sample(5000)
print('Number of Rows after taking sampling :',df.shape[0]) #Row s hain 
print('Number of Columns after taking sampling:',df.shape[1]) # ya columns hain 
```

```python
df.head()
```

# set Title 

```python
sns.scatterplot(x='carat',y='price',data=df)
```

```python
sns.scatterplot(x='carat',y='price',data=df,hue='cut')
plt.title('Price vs Carat')
plt.show()
```

# setting x-labels 

```python
sns.scatterplot(x='carat',y='price',data=df,hue='cut')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

# Customize the legends 

```python
sns.scatterplot(x='carat',y='price',data=df,hue='cut')
plt.legend(title='Diamond cut',loc='upper right')
plt.show()
```

# x-ticks and y-ticks

```python
sns.histplot(x='price',data=df,bins=10,edgecolor='black',hue='cut')
plt.xticks(rotation=45)
plt.yticks() # No need to rotate it 
plt.show()
```

# Adding a grid

```python
sns.scatterplot(x='carat',y='price',data=df,hue='cut')
sns.set_style('darkgrid')
plt.grid(linestyle='--')
plt.show()
```

# Customizing spines 

```python
sns.set_theme(style='darkgrid')
sns.violinplot(x='cut',y='price',data=df,hue='color')
plt.grid(linestyle='--')
sns.despine()
plt.show()
```

# figure size 

```python
figure=plt.figure(figsize=(10,6))
sns.barplot(x='cut',y='price',data=df,hue='color')
plt.title('Price vs Cut with large figure size')
plt.show()
```

# Creting the Subplots 

```python
fig,axs=plt.subplots(1,2,figsize=(15,8))
sns.barplot(x='cut',y='price',data=df,hue='color',ax=axs[0])
sns.scatterplot(x='carat',y='price',data=df,hue='color',ax=axs[1])
plt.show()
```

# Color palette

```python
figure=plt.figure(figsize=(15,8))
sns.set_palette("colorblind")  # you can change the colors of the graph like deep,muted,pastel,bright,dark and colorblind 
sns.barplot(x='cut',y='price',data=df,hue='color')
plt.title('Barplot of price vs cut with colorblind palette colors')
plt.show()
```
