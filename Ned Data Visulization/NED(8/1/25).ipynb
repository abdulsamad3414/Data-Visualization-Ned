```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
digits=load_digits(n_class=6)

fig,ax=plt.subplots(8,8,figsize=(6,6))
for i,axi in enumerate(ax.flat):
    axi.imshow(digits.images[i],cmap='Greens') # it ca be change 
    axi.set(xticks=[],yticks=[])
digits.data
```

```python
digits.target.shape
```

```python
print('images data in pixel',digits.data[99])
print('actually it is :',digits.target[99])
```

```python
plt.imshow(digits.images[99],cmap='binary')
print('images data in pixel',digits.data[55])
print('actually it is :',digits.target[55])
```

```python
plt.imshow(digits.images[99],cmap='Greens')
print('images data in pixel',digits.data[4])
print('actually it is :',digits.target[5])
```

```python
# Project the digit into 2 dimension using ISOMAP
from sklearn.manifold import Isomap

# Create an instance of Isomap and fit it to the data
iso=Isomap(n_components=2)
projection=iso.fit_transform(digits.data)
```

```python
# Plots the Results 
# plt.figure(figsize=(20,10))
plt.scatter(projection[:,0],projection[:,1],lw=0.1,
            c=digits.target,cmap='cubehelix')
plt.colorbar(ticks=range(6),label='digit value')
plt.clim(-0.5,5.5) # this is color limits 
```
